```python
import pandas as pd

df = pd.read_csv('./peso_altura_.csv')
datos = df 
datos 
```

```python
import seaborn as sb 
sb.scatterplot(x= "altura",y = "peso", data = datos, hue ="peso", palette = "coolwarm")
```

```python
#se preparan los datos para entrenamiento
#caracteristica (X), etiqueta (y)

X = datos["altura"]
y = datos["peso"]
```

```python
X_transform = X.values.reshape(-1,1)
y_transform = y.values.reshape(-1,1)
 #cuentan con el formato especifico
```

```python
from sklearn.linear_model import LinearRegression

modelo_altura_peso = LinearRegression()

```

```python
modelo_altura_peso.fit(X_transform, y_transform)
```

```python
import numpy as np

altura = np.array([[1.67]])

peso_prediccion = modelo_altura_peso.predict(altura)
peso_prediccion 
```

```python
modelo_altura_peso.score(X_transform, y_transform)
```
